```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house2 = pd.read_csv('CLEAN_House2.csv', nrows = 500000)
house2.head()
```

```python
house2.describe()
```

```python
house2.columns
```

```python
subset = house2.iloc[:10000]

plt.figure(figsize=(15,8))
plt.plot(subset.index , subset["Aggregate"] , label = "Aggregate power")
plt.plot(subset.index , subset["Appliance1"], label = "Fridge Freezer")
plt.xlabel("Time")
plt.ylabel("Power (W)")
plt.legend()
plt.title("Power Consumption Over Time")
plt.tight_layout()
plt.show()
```

```python
subset = house2.iloc[8000:10000]

plt.figure(figsize=(15,8))
plt.plot(subset.index , subset["Aggregate"] , label = "Aggregate power")
plt.plot(subset.index , subset["Appliance3"], label = "Dishwasher")
plt.xlabel("Time")
plt.ylabel("Power (W)")
plt.legend()
plt.title("Power Consumption Over Time")
plt.tight_layout()
plt.show()
```
